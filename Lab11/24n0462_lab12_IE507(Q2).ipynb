```python
#we can also use annotations for bar plots
#We will consider the bar plot based on gender vs serum_cholesterol
plot = sns.barplot(x="gender",y="serum_cholesterol",data=df)


for p,line in zip(plot.patches, plot.lines):
    plot.annotate(format(p.get_height(),'.1f'),
                   (p.get_x() + p.get_width() / 2., line.get_ydata()[1]),
                   ha = 'center', va = 'top',
                   xytext = (0, 9),
                   textcoords = 'offset points')
#Try to find the meaning of the different arguments used in annotate function
```

Q6

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
bin = [ 20, 50, 100, 200, 500]

for bin_size in bin:
  print(f"Bin size: {bin_size}")
  # Pandas histogram
  df.hist(column='serum_cholesterol', bins=bin_size)
  plt.title(f'Pandas Histogram (bins={bin_size})')
  plt.show()

  # Seaborn histogram with KDE
  sns.histplot(df.serum_cholesterol, bins=bin, kde=True)
  plt.title(f'Seaborn Histogram with KDE (bins={bin_size})')
  plt.show()

  print("-" * 20)
```
