```python
# inbound trains from C1 to M1 (in hours, for simplicity)
c1_to_m1_arrival = [5.50, 10.25, 12.30, 18.15, 21.05, 16.20]
```

```python
# outbound train departure times from M1 to S1 and S2
m1_to_s1 = [18.25, 6.20, 2.25, 5.15, 9.20, 15.15, 20.18, 22.20]
m1_to_s2 = [12.25, 15.16, 18.21, 7.25, 21.20]
```

```python
# Minimum layover time at M1 is 1 hour
min_layover = 1
```

```python
# Function to find and print feasible trains
def print_feasible_trains():
    for i, arrival in enumerate(c1_to_m1_arrival):
        print(f"\nInbound Train {i+1} (Arrival from C1 to M1 at {arrival} hours):")

        # Find feasible outbound trains from M1 to S1
        feasible_s1_trains = [dep for dep in m1_to_s1 if dep > arrival + min_layover]
        print(f"  Feasible outbound trains from M1 to S1 (with layover >= {min_layover} hours): {feasible_s1_trains}")

        # Find feasible outbound trains from M1 to S2
        feasible_s2_trains = [dep for dep in m1_to_s2 if dep > arrival + min_layover]
        print(f"  Feasible outbound trains from M1 to S2 (with layover >= {min_layover} hours): {feasible_s2_trains}")

# Call the function to print feasible trains
print_feasible_trains()
```

```python
m1_to_s2_travel = [12.25, 18.21, 21.20, 7.25, 15.16]
m1_to_s2_departure = [12.25, 15.16, 18.21, 7.25, 21.20]
```

```python
min_layover = 1
```

```python
feasible_plans = []

for i, arrival_time in enumerate(c1_to_m1_arrival):
    for s1_dep, s1_travel in zip(m1_to_s1_departure, m1_to_s1_travel):
        for s2_dep, s2_travel in zip(m1_to_s2_departure, m1_to_s2_travel):
            if s1_dep > arrival_time + min_layover and s2_dep > arrival_time + min_layover:
                # Calculate waiting time
                waiting_time = abs(s2_dep - s1_dep)
                # Total travel time
                total_travel_time = 2 * c1_to_m1_travel[i] + s1_travel + s2_travel + waiting_time + min_layover
                feasible_plans.append((i + 1, s1_dep, s2_dep, total_travel_time))
```

```python
layover_times = [1, 2]
max_waiting_times = [2, 5, 10]
```

```python
def find_feasible_plans(layover, max_wait):
    feasible_plans = []
    for i, arrival_time in enumerate(c1_to_m1_arrival):
        for s1_dep, s1_travel in zip(m1_to_s1_departure, m1_to_s1_travel):
            for s2_dep, s2_travel in zip(m1_to_s2_departure, m1_to_s2_travel):
                if s1_dep > arrival_time + layover and s2_dep > arrival_time + layover:
                    # Calculate waiting time (if second passenger departs later)
                    waiting_time = abs(s2_dep - s1_dep)
                    if waiting_time <= max_wait:
                        # Total travel time
                        total_travel_time = 2 * c1_to_m1_travel[i] + s1_travel + s2_travel + waiting_time + layover
                        feasible_plans.append((i + 1, s1_dep, s2_dep, total_travel_time))

    # Sort the plans by total travel time
    feasible_plans.sort(key=lambda x: x[3])
    return feasible_plans
```

```python
for layover in layover_times:
    for max_wait in max_waiting_times:
        print(f"\nLayover Time = {layover} hours || Maximum Waiting Time = {max_wait} hours")
        feasible_plans = find_feasible_plans(layover, max_wait)
        if feasible_plans:
            print("TOP 3 FEASIBLE TRAVEL PLANS")
            for plan in feasible_plans[:3]:
                print(f"From C1 to M1 take train {plan[0]} |from M1 to S1 departs at {plan[1]}| "
                      f"from M1 to S2 departs at {plan[2]} | Total travel time = {plan[3]:.2f} hours")
        else:
            print("NO PLAN FOUND")
```

```python
# Train travel times
c1_to_m1_travel = [3.42, 2.56, 4.15, 3.18, 5.08, 3.50]
c1_to_m1_arrival = [5.50, 10.25, 12.30, 18.15, 21.05, 16.20]
```

```python
m1_to_s1_travel = [2.25, 5.15, 9.20, 6.20, 15.15, 18.25]
m1_to_s1_departure = [18.25, 6.20, 2.25, 5.15, 9.20, 15.15]
```

```python
m1_to_s2_travel = [12.25, 18.21, 21.20, 7.25, 15.16]
m1_to_s2_departure = [12.25, 15.16, 18.21, 7.25, 21.20]
```

```python
def top_5_plans():
    travel_plans = []

    for i, arrival_1 in enumerate(c1_to_m1_arrival):
        for j, arrival_2 in enumerate(c1_to_m1_arrival):
            for s1_dep, s1_travel in zip(m1_to_s1_departure, m1_to_s1_travel):
                for s2_dep, s2_travel in zip(m1_to_s2_departure, m1_to_s2_travel):
                    # Calculate waiting times
                    wait_1 = s1_dep - arrival_1 if s1_dep > arrival_1 else 0
                    wait_2 = s2_dep - arrival_2 if s2_dep > arrival_2 else 0


                    total_time = (c1_to_m1_travel[i] + s1_travel + c1_to_m1_travel[j] + s2_travel + wait_1 + wait_2)

                    # Store the travel plan
                    travel_plans.append((i+1, j+1, s1_dep, s2_dep, total_time))

    travel_plans.sort(key=lambda x: x[4])
    # Print top 5 travel plans
    print("Top 5 feasible travel plans:")
    for plan in travel_plans[:5]:
        print(f"Passenger 1: From C1 to M1  {plan[0]}",end = " ")
        print(f"from M1 to S1 departs at {plan[2]}. ")
        print(f"Passenger 2: From C1 to M1 {plan[1]}; ",end = " ")
        print(f"from M1 to S2 departs at {plan[3]}. ")
        print(f"Total travel time = {plan[4]:.2f} hours")


# Call the function to find top 5 plans
top_5_plans()
```

```python
# Train travel times
c1_to_m1_travel = [3.42, 2.56, 4.15, 3.18, 5.08, 3.50]
c1_to_m1_arrival = [5.50, 10.25, 12.30, 18.15, 21.05, 16.20]

m1_to_s1_travel = [2.25, 5.15, 9.20, 6.20, 15.15, 18.25]
m1_to_s1_departure = [18.25, 6.20, 2.25, 5.15, 9.20, 15.15]

m1_to_s2_travel= [12.25, 18.21, 21.20, 7.25, 15.16]
m1_to_s2_departure = [12.25, 15.16, 18.21, 7.25, 21.20]
```

```python
def top_5_plans_max_wait():
    travel_plans = []


    for i, arrival_1 in enumerate(c1_to_m1_arrival):
        for j, arrival_2 in enumerate(c1_to_m1_arrival):
            # Check the waiting time at C1
            wait_c1 = abs(c1_to_m1_arrival[j] - c1_to_m1_arrival[i])

            if wait_c1 >= 10:
                continue

            for s1_dep, s1_travel in zip(m1_to_s1_departure, m1_to_s1_travel):
                for s2_dep, s2_travel in zip(m1_to_s2_departure, m1_to_s2_travel):
                    # Calculate waiting times at M1
                    wait_m1_1 = s1_dep - arrival_1 if s1_dep > arrival_1 else 0
                    wait_m1_2 = s2_dep - arrival_2 if s2_dep > arrival_2 else 0

                    if wait_m1_1 >= 10 or wait_m1_2 >= 10:
                        continue


                    total_time = (c1_to_m1_travel[i] + s1_travel + c1_to_m1_travel[j] + s2_travel + wait_c1 + wait_m1_1 + wait_m1_2)

                    # Store the travel plan
                    travel_plans.append((i+1, j+1, s1_dep, s2_dep, total_time))

    travel_plans.sort(key=lambda x: x[4])

    # Print top 5 travel plans
    print("TOP 5 FEASIBLE PLANS WITH TIME <10 \n")
    for plan in travel_plans[:5]:
        print(f"Passenger 1: From C1 to M1  {plan[0]}||"
              f"from M1 to S1 departs at {plan[2]}. "
              f"Passenger 2: From C1 to M1  {plan[1]}|| "
              f"from M1 to S2 departs at {plan[3]}. "
              f"Total travel time = {plan[4]:.2f} hours\n")

# Calling the function
top_5_plans_max_wait()
```

```python
def top_5_plans_with_conditions():
    travel_plans = []


    for i, arrival_1 in enumerate(c1_to_m1_arrival):
        for j, arrival_2 in enumerate(c1_to_m1_arrival):
            for s1_dep, s1_travel in zip(m1_to_s1_departure, m1_to_s1_travel):
                for s2_dep, s2_travel in zip(m1_to_s2_departure, m1_to_s2_travel):
                    # Calculate waiting times at M1
                    wait_m1_1 = s1_dep - arrival_1 if s1_dep > arrival_1 else 0
                    wait_m1_2 = s2_dep - arrival_2 if s2_dep > arrival_2 else 0
                    # Calculate waiting times at C1
                    wait_c1 = abs(c1_to_m1_arrival[j] - c1_to_m1_arrival[i])

                    # Total waiting times
                    W1 = wait_c1 + wait_m1_1
                    W2 = wait_c1 + wait_m1_2


                    if W1 >= max(5, 2 * W2) or W2 >= max(5, 2 * W1):
                        continue  # Skip if the condition is violated

                    # Total travel time for both passengers
                    total_time = (c1_to_m1_travel[i] + s1_travel +
                                  c1_to_m1_travel[j] + s2_travel +
                                  wait_m1_1 + wait_m1_2 + wait_c1)

                    # Store the travel plan
                    travel_plans.append((i+1, j+1, s1_dep, s2_dep, total_time, W1, W2))

    # Sort the travel plans by total travel time
    travel_plans.sort(key=lambda x: x[4])

    # Print top 5 travel plans
    print("Top 5 feasible travel plans with W1 and W2 conditions:")
    for plan in travel_plans[:5]:
        print(f"  - Passenger 1: From C1 to M1 take train {plan[0]}",end = "")
        print(f"    Departure from M1 to S1: {plan[2]} hours")
        print(f"  - Passenger 2: From C1 to M1 take train {plan[1]}",end = "")
        print(f"    Departure from M1 to S2: {plan[3]} hours")
        print(f"  - Total travel time: {plan[4]:.2f} hours")
        print(f"  - Waiting time W1: {plan[5]:.2f} hours")
        print(f"  - Waiting time W2: {plan[6]:.2f} hours\n")


top_5_plans_with_conditions()
```
