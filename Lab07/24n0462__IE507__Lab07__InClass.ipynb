```python
from pyomo.environ import*
```

```python
model = ConcreteModel()
```

```python
#decentralised
# declaring decision variable for system
model.x1 = Var(domain = Reals)
model.x2 = Var(domain = Reals)
model.y1 = Var(domain = Reals)
model.y2 = Var(domain = Reals)

# declaring constraints
model.constr1 = Constraint(expr = model.x1 >= 0)
model.constr2 = Constraint(expr = model.x2 >= 0)
model.constr3 = Constraint(expr = model.y1 >= 0)
model.constr4 = Constraint(expr = model.y2 >= 0)
model.constr5 = Constraint(expr = 4*model.x1 + 2*model.x2 <= 80)
model.constr6 = Constraint(expr = 2*model.x1 + 5*model.x2 <= 60)
model.constr7 = Constraint(expr = 5*model.y1 + 3*model.y2 <= 60)
model.constr8 = Constraint(expr = 5*model.y1 + 6*model.y2 <= 75)
model.constr9 = Constraint(expr = model.x1 + model.x2 <= 60)
model.constr10 = Constraint(expr = model.y1 + model.y2 <= 60)

# declaring objective function for optimization problem
model.cost = Objective(expr = 10*model.x1 + 15*model.x2 + 10*model.y1 + 15*model.y2, sense=maximize)

```

```python
!apt-get install -y -qq glpk-utils
SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()
SolverFactory()
```

```python
print('\nCost = ', model.cost())
print('\nDecision Variables')
print('x1 = ', model.x1())
print('x2 = ', model.x2())
print('y1 = ', model.y1())
print('y2 = ', model.y2())

print('\nConstraints')
print('Constraint 1 = ', model.constr1())
print('Constraint 2 = ', model.constr2())
print('Constraint 3 = ', model.constr3())
print('Constraint 4 = ', model.constr4())
print('Constraint 5 = ', model.constr1())
print('Constraint 6 = ', model.constr2())
print('Constraint 7 = ', model.constr3())
print('Constraint 8 = ', model.constr4())
print('Constraint 9 = ', model.constr1())
print('Constraint 10 = ', model.constr2())
```
