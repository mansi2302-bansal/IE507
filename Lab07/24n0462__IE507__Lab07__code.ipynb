```python
model = ConcreteModel()
```

```python
#decentralised
# declaring decision variable for system
model.x1 = Var(domain = Reals)
model.x2 = Var(domain = Reals)
model.y1 = Var(domain = Reals)
model.y2 = Var(domain = Reals)

# declaring constraints
model.constr1 = Constraint(expr = model.x1 >= 0)
model.constr2 = Constraint(expr = model.x2 >= 0)
model.constr3 = Constraint(expr = model.y1 >= 0)
model.constr4 = Constraint(expr = model.y2 >= 0)
model.constr5 = Constraint(expr = 4*model.x1 + 2*model.x2 <= 80)
model.constr6 = Constraint(expr = 2*model.x1 + 5*model.x2 <= 60)
model.constr7 = Constraint(expr = 5*model.y1 + 3*model.y2 <= 60)
model.constr8 = Constraint(expr = 5*model.y1 + 6*model.y2 <= 75)
model.constr9 = Constraint(expr = model.x1 + model.x2 <= 60)
model.constr10 = Constraint(expr = model.y1 + model.y2 <= 60)

# declaring objective function for optimization problem
model.cost = Objective(expr = 10*model.x1 + 15*model.x2 + 10*model.y1 + 15*model.y2, sense=maximize)

```

```python
!apt-get install -y -qq glpk-utils
SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()
SolverFactory()
```

```python
model2 = ConcreteModel()
```

```python
# Allocation
# declaring decision variable for system
import numpy as np
import matplotlib.pyplot as plt

def decentr_solve(r):
  model2.x1 = Var(domain = Reals)
  model2.x2 = Var(domain = Reals)
  model2.y1 = Var(domain = Reals)
  model2.y2 = Var(domain = Reals)

  #constraints
  model2.constr1 = Constraint(expr = model2.x1 >= 0)
  model2.constr2 = Constraint(expr = model2.x2 >= 0)
  model2.constr3 = Constraint(expr = model2.y1 >= 0)
  model2.constr4 = Constraint(expr = model2.y2 >= 0)
  model2.constr5 = Constraint(expr = 4*model2.x1 + 2*model2.x2 <= 80)
  model2.constr6 = Constraint(expr = 2*model2.x1 + 5*model2.x2 <= 60)
  model2.constr7 = Constraint(expr = 5*model2.y1 + 3*model2.y2 <= 60)
  model2.constr8 = Constraint(expr = 5*model2.y1 + 6*model2.y2 <= 75)
  model2.constr9 = Constraint(expr = 5*model2.y1 + 6*model2.y2 <= 75)
  model2.constr10 = Constraint(expr = model2.x1 + model2.x2 <= 60*r )
  model2.constr11 = Constraint(expr = model2.y1 + model2.y2 <= 60*(1-r))

  # declaring objective function for optimization problem
  model2.cost = Objective(expr = 10*model2.x1 + 15*model2.x2 + 10*model2.y1 + 15*model2.y2, sense=maximize)

  SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model2).write()
  SolverFactory()


  return value(model2.cost)

r_values = np.linspace(0,1,100)
profit = []
for r in r_values:
  output = decentr_solve(r)
  profit.append(output)

plt.plot(r_values, profit)
plt.xlabel('fraction of crude oil (r)')
plt.ylabel('profit')
plt.show()

```
